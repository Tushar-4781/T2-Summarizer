```python
# Perform standard imports
import spacy
from spacy.lang.en.stop_words import STOP_WORDS
from string import punctuation
nlp = spacy.load('en_core_web_sm')

def summarizer(text):
    # Write a function to display basic entity info:
    def show_ents(doc):
        #print("entities\n")
        enty=[]
        if doc.ents:
            for ent in doc.ents:
                if ent.text not in enty:
                    #print(ent.text)#+' - '+ent.label_+' - '+str(spacy.explain(ent.label_)))
                    enty.append(ent.text)
        return enty
    
    # summary points
    summary=[]
    
    doc = nlp(text)
    
    # list of entities
    enty = show_ents(doc)
    
    # stop words, tokenization, punctuation
    stopwords = list(STOP_WORDS)
    tokens = [token.text for token in doc]
    punctuation='!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~:\n\n'

    # Generating a dictionary of word frequencies
    word_frequencies = {}
    for word in doc:
        if word.text.lower() not in stopwords:
            if word.text.lower() not in punctuation:
                if word.text not in word_frequencies.keys():
                    word_frequencies[word.text] = 1
                else:
                    word_frequencies[word.text] += 1
    #print(word_frequencies)

    max_frequency = max(word_frequencies.values())
    max_frequency

    for word in word_frequencies.keys():
        word_frequencies[word] = word_frequencies[word]/max_frequency
    #print(word_frequencies)
    
    # sentence tokens
    sentence_tokens = [sent for sent in doc.sents]
    #print(sentence_tokens)
    
    # adding the first statement by default in summary
    summary.append(sentence_tokens[0])
    
    # Generating a dictionary of sentence Scores
    sentence_scores = {}
    for sent in sentence_tokens:
        for word in sent:
            if word.text.lower() in word_frequencies.keys():
                if sent not in sentence_scores.keys():
                    sentence_scores[sent] = word_frequencies[word.text.lower()]
                else:
                    sentence_scores[sent] += word_frequencies[word.text.lower()]
                    if word.text.lower() in enty:
                        sentence_scores[sent] += word_frequencies[word.text.lower()]
    #print(sentence_scores)
    #print("total sentences "+str(len(sentence_scores)))
    
    # % of summarization to be done
    upto = int(input("Enter the value for reduction like 50 for 50%: "))/100
    
    # calculating window for selecting the sentecnes
    period = int(len(sentence_scores)*(upto**2))
    #print("interval : ",period)
    
    # generating Summary
    cnt=0
    mx_freq=0
    sent=""
    for i in sentence_scores:
        cnt+=1
        if(sentence_scores[i]>mx_freq):
            sent=i
            mx_freq=sentence_scores[i]
        if(cnt==period):
            cnt=0
            mx_freq=0
            if(sent not in summary):
                summary.append(sent)
                
    final_summary = [word.text for word in summary]
    final_summary = ' '.join(final_summary)
    #for i in sentence_scores:
    #    print(i,sentence_scores[i])
    # if final sentence not in summary add it
    if str(list(sentence_scores.keys())[-1]) not in final_summary:
        final_summary+=str(list(sentence_scores.keys())[-1])
    print("\nMy SUMMARY:\n")
    print(final_summary)
    print("\nreduce from "+str(len(text))+" to "+str(len(final_summary)))
```
